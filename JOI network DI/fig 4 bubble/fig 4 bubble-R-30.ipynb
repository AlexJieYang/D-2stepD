```r
library(reshape2)
library(ggplot2)
```

# APS

```r
df<-read.csv('E:/netData/complexNetdataset/bubbleMatrixData/APStopp30.csv',header = T,encoding = "UTF-8")
```

```r
names(df) <- c("X", 'D', '2stepD', 'CD', 'degree' ,'h' ,'PR')
# define the desired order of the X factor variable
x_order <- c("D", "2stepD", "CD", "degree", "h", "pagerank")

# reorder the X factor variable in the original data frame
df$X <- factor(df$X, levels = x_order)

# melt the data
df_melt <- melt(df, id.vars = "X", variable.name = "variable", value.name = "value")

# convert to percentage and round to two decimal places
df_melt$value <- round(df_melt$value * 100, 2)

# create the ggplot2 plot with text labels
ggplot(df_melt, aes(x = variable, y = X, size = value, alpha = value)) +
  geom_point(color = "firebrick", shape = "diamond") +
  geom_text(aes(label = paste0(value, "%")), size = 4.5) +
  scale_size_continuous(range = c(15, 30)) +
  theme(
    panel.background = element_blank(),
    panel.grid = element_line("gray"),
    panel.border = element_rect(colour = "black", fill = NA),
    axis.text = element_text(color = "black", size = 21),
    legend.position = "none",
    panel.grid.major = element_line(color = "#E6B0AA", linetype = "dashed", size = .5)
  ) +
  labs(x = NULL, y = "") +
  scale_alpha_continuous(range = c(0.5, 1)) +
  coord_fixed()
ggsave("F:/A J YANG PAPER/Disruption index/JOI DI in complex networks/fig/fig4/APS30.pdf")
```
